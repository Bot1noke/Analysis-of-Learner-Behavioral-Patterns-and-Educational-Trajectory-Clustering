### 1. Фичи успешности (Базовые)
* **accuracy**: доля правильных решений (среднее значение is_correct по всем попыткам пользователя)
* **avg_normalized_score**: средний нормированный балл за попытки (score / max_score)

> Я нормализовал score, так как задания имеют разный максимальный балл.

### 2. Фичи активности (Базовые)
* **submissions_count**: общее количество попыток решения задач (число строк в логах пользователя, каждая строка = одна отправка решения)
* **unique_tasks**: количество уникальных задач, которые пользователь решал
* **unique_practices**: количество уникальных практик (practice_submission_id), в которых участвовал пользователь
* **active_days**: количество дней, в которые пользователь делал хотя бы одну отправку

### 3. Поведенческие фичи (Базовые)
* **repeat_rate**: доля повторных попыток (1 − unique_tasks / submissions_count), показывает, как часто пользователь переотправляет решения одних и тех же задач
* **submissions_per_day**: среднее число отправок в активный день (submissions_count / active_days)
* **tasks_per_practice**: среднее количество уникальных задач на одну практику (unique_tasks / unique_practices)

### 4. Тематические фичи
* **unique_topics**: количество уникальных компетенций (тем), с которыми взаимодействовал пользователь
* **topic_entropy**: мера разнообразия тем (чем выше, тем более равномерно пользователь работает с разными темами)

> Энтропия показывает, изучает ли пользователь темы равномерно или сосредоточен на каком-то узком наборе тем/задач.

```python
df_raw["normalized_score"] = df_raw["score"] / df_raw["max_score"]

user_features = df_raw.groupby("user_number").agg(
    accuracy=("is_correct", "mean"),
    submissions_count=("submission_id", "count"),
    unique_tasks=("task_id", "nunique"),
    avg_normalized_score=("normalized_score", "mean"),
    unique_practices=("practice_submission_id", "nunique")
)
```

```python
df_raw["submission_time"] = pd.to_datetime(df_raw["submission_time"], errors="coerce")
df_raw["submission_date"] = df_raw["submission_time"].dt.date

user_features["active_days"] = df_raw.groupby("user_number")["submission_date"].nunique()
user_features["repeat_rate"] = 1 - user_features["unique_tasks"] / user_features["submissions_count"]
user_features["submissions_per_day"] = (
    user_features["submissions_count"] / user_features["active_days"].replace(0, np.nan)
)
user_features["tasks_per_practice"] = (
    user_features["unique_tasks"] / user_features["unique_practices"]
)
```

```python
df_topic = df_raw.merge(mapping, on="task_id", how="left")

def entropy(x):
    x = x.dropna()
    if len(x) == 0:
        return np.nan
    probs = x.value_counts(normalize=True)
    return -(probs * np.log(probs + 1e-9)).sum()

user_features["topic_entropy"] = df_topic.groupby("user_number")["competence_id"].apply(entropy)
user_features["unique_topics"] = df_topic.groupby("user_number")["competence_id"].nunique(dropna=True)
```

```python
user_features = user_features.reset_index()

print(user_features.head())
print()
print(user_features.describe())
```

```python
print("Пропуски по признакам:")
print(user_features.isna().sum())
print()

print("Пользователи с самой высокой активностью:")
print(user_features.sort_values("submissions_count", ascending=False)[
        ["user_number", "submissions_count", "unique_tasks", "active_days", "repeat_rate"]
    ].head(10))
print()
print("Пользователи с самой высокой accuracy:")
print(user_features.sort_values("accuracy", ascending=False)[
        ["user_number", "accuracy", "submissions_count", "unique_tasks"]
    ].head(10))
```
