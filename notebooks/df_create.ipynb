```python
import pandas as pd
import json
import re
from html import unescape
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
with open('fipi_exported_katex.json', 'r', encoding='utf-8') as f:
    fipi_data = json.load(f)

with open('all_variants_all_years.json', 'r', encoding='utf-8') as f:
    real_data = json.load(f)
```

```python

def clean_html(text):
    if text is None:
        return None
    text = str(text)
    text = unescape(text)
    text = re.sub(r'<[^>]+>', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

```

```python

fipi_rows = []

for task in fipi_data:

    problem_src = task.get('problem_katex') or task.get('problem')

    fipi_rows.append({
        'dataset_source': 'fipi',
        'task_id': task.get('id'),
        'task_number': task.get('task_number'),
        'question_text': clean_html(task.get('question')),
        'problem_text': clean_html(problem_src),
        'question_raw': task.get('question'),
        'problem_raw': problem_src,
        'codifier': task.get('codifier'),
        'answer_type': task.get('answer_type'),
        'answer': task.get('answer')
    })

df_fipi = pd.DataFrame(fipi_rows)
```

```python
df_fipi.head()
```

```python

real_rows = []

for year_block, variants in real_data.items():
    for variant in variants:
        variant_title = variant.get('title')
        variant_url = variant.get('url')

        for task in variant.get('tasks', []):
            real_rows.append({
                'dataset_source': 'real_exam',
                'year_block': year_block,
                'variant_title': variant_title,
                'variant_url': variant_url,
                'variant_id': task.get('source_id') or variant_title,

                'task_id': task.get('id'),
                'task_number': task.get('task_number'),

                'question_raw': task.get('question'),
                'problem_raw': task.get('problem'),

                'question_text': clean_html(task.get('question')),
                'problem_text': clean_html(task.get('problem')),

                'codifier': task.get('codifier'),
                'answer_type': task.get('answer_type'),
                'answer': task.get('answer')
            })

df_real = pd.DataFrame(real_rows)
```

```python


def extract_year(title):
    if not title:
        return None
    m = re.search(r'\d{2}\.\d{2}\.(\d{4})', title)
    if m:
        return int(m.group(1))
    m = re.search(r'[Nn]?(\d{4})', title)
    if m:
        return int(m.group(1))
    return None

rows = []

files = {
    'result1.json': 'shkolkovo_fipi',
    'result2.json': 'shkolkovo_fipi',
    'result3.json': 'shkolkovo_fipi',
    'result4.json': 'shkolkovo_author',
}

for filename, source in files.items():
    with open(filename) as f:
        data = json.load(f)

    for variant in data:
        if 'error' in variant or 'questions' not in variant:
            continue

        url        = variant.get('url')
        title      = variant.get('title')
        year_block = extract_year(title)

        for q in variant['questions']:
            text = q.get('text')
            rows.append({
                'dataset_source': source,
                'task_id':        q.get('exercise_id'),
                'task_number':    q.get('number'),
                'question':       text,
                'problem':        text,
                'question_raw':   text,
                'problem_raw':    text,
                'answer':         q.get('answer'),
                'solution':       q.get('solution'),
                'answer_options': None,
                'variant_url':    url,
                'variant_title':  title,
                'year_block':     year_block,
            })

df_new = pd.DataFrame(rows)
print(f'Новых строк: {len(df_new)}')
print(df_new['dataset_source'].value_counts())


df_main = pd.read_csv('tasks_with_oge.csv')
print(f'\nОсновной датасет: {len(df_main)} строк')

df_merged = pd.concat([df_main, df_new], ignore_index=True, sort=False)
print(f'Итого после склейки: {len(df_merged)} строк')
print(df_merged['dataset_source'].value_counts())


df_merged.to_csv('tasks_merged.csv', index=False)
print('\nСохранено в tasks_merged.csv')
```
